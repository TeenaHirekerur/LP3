```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import ConfusionMatrixDisplay,confusion_matrix, accuracy_score, precision_score, recall_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
df=pd.read_csv("C:\\Users\\Teena Hirekerur\\Downloads\\archive (7)\\emails.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
x = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train, y_train)
knn_pred = knn.predict(x_test)
ConfusionMatrixDisplay.from_estimator(knn, x_test, y_test)
```

```python
svm = SVC(gamma='auto', random_state=10)
svm.fit(x_train, y_train)
svm_pred = svm.predict(x_test)
ConfusionMatrixDisplay.from_estimator(svm, x_test, y_test)
```

```python
accuracy_score(y_test,svm_pred)
```

```python
confusion_matrix(y_test,svm_pred)
```
